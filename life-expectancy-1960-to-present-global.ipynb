```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
height = 800
kaggle = True

if kaggle:
    data_path = '/kaggle/input/life-expectancy-1960-to-present-global/life_expectancy_dataset.csv'
else:
    data_path = './data/life_expectancy_dataset.csv'
df_data = pd.read_csv(data_path, sep=';', decimal=',')
df_data.head()
```

```python
df_data = df_data[(df_data['life_expectancy_women'] > 0) | (df_data['life_expectancy_men'] > 0)]
df_data.describe()
```

```python
# Δημιουργία γραφήματος με δύο υποδιαγράμματα
fig = make_subplots(rows=1, 
                    cols=2, 
                    subplot_titles=("Προσδόκιμο Ζωής Γυναικών", "Προσδόκιμο Ζωής Ανδρών"))

# Προσθήκη ιστογράμματος για τις γυναίκες
fig.add_trace(go.Histogram(x=df_data["life_expectancy_women"],
                           nbinsx=30, 
                           name="Γυναίκες", 
                           marker_color="salmon"), 
              row=1, col=1)

# Προσθήκη ιστογράμματος για τους άνδρες
fig.add_trace(go.Histogram(x=df_data["life_expectancy_men"], 
                           nbinsx=30, 
                           name="Άνδρες",
                           marker_color="skyblue"), 
              row=1, col=2)

# Ρύθμιση του layout για καλύτερη εμφάνιση
fig.update_layout(
    title_text="Κατανομή Προσδόκιμου Ζωής",  # Τίτλος γραφήματος
    showlegend=False,  # Απενεργοποίηση της υπογραφής
    height=600,  # Ύψος γραφήματος
    width=1000,  # Πλάτος γραφήματος
    plot_bgcolor='rgba(0,0,0,0)',  # Διαφάνεια φόντου
    font=dict(family="Arial, sans-serif", size=12, color="black"),  # Ρύθμιση γραμματοσειράς
)

# Ρύθμιση για άξονες X
fig.update_xaxes(title_text="Προσδόκιμο Ζωής (Έτη)", row=1, col=1)
fig.update_xaxes(title_text="Προσδόκιμο Ζωής (Έτη)", row=1, col=2)

# Ρύθμιση για άξονες Y
fig.update_yaxes(title_text="Συχνότητα", row=1, col=1)
fig.update_yaxes(title_text="Συχνότητα", row=1, col=2)

# Εμφάνιση του γραφήματος
fig.show()
```

```python
# Obtener el año más reciente disponible en el dataset
latest_year = df_data["year"].max()

# Filtrar los datos del año más reciente
df_latest = df_data[df_data["year"] == latest_year]

# Obtener los 5 países con mayor y menor esperanza de vida para mujeres
top_countries_women = df_latest.nlargest(5, "life_expectancy_women")
bottom_countries_women = df_latest.nsmallest(5, "life_expectancy_women")

# Obtener los 5 países con mayor y menor esperanza de vida para hombres
top_countries_men = df_latest.nlargest(5, "life_expectancy_men")
bottom_countries_men = df_latest.nsmallest(5, "life_expectancy_men")
```

```python
import plotly.graph_objects as go

# Δημιουργία γραφήματος για το προσδόκιμο ζωής των γυναικών
fig_women = go.Figure()

# Κορυφαία 5 χώρες με το μεγαλύτερο προσδόκιμο ζωής για γυναίκες
fig_women.add_trace(go.Bar(x=top_countries_women["country_name"],
                           y=top_countries_women["life_expectancy_women"],
                           name="Κορυφαία 5 χώρες για γυναίκες",
                           marker_color="salmon"))

# Κορυφαία 5 χώρες με το μικρότερο προσδόκιμο ζωής για γυναίκες
fig_women.add_trace(go.Bar(x=bottom_countries_women["country_name"],
                           y=bottom_countries_women["life_expectancy_women"],
                           name="Χώρες με το μικρότερο προσδόκιμο ζωής για γυναίκες",
                           marker_color="lightcoral"))

# Ρύθμιση τίτλου και άξονων
fig_women.update_layout(title="Προσδόκιμο Ζωής Γυναικών - Κορυφαία και Χαμηλότερα Κράτη",
                        xaxis_title="Χώρες",
                        yaxis_title="Προσδόκιμο Ζωής (Έτη)",
                        barmode="group",  # Ομαδοποίηση των μπαρ
                        height=600,
                        width=900)

fig_women.show()

# Δημιουργία γραφήματος για το προσδόκιμο ζωής των ανδρών
fig_men = go.Figure()

# Κορυφαία 5 χώρες με το μεγαλύτερο προσδόκιμο ζωής για άνδρες
fig_men.add_trace(go.Bar(x=top_countries_men["country_name"],
                         y=top_countries_men["life_expectancy_men"],
                         name="Κορυφαία 5 χώρες για άνδρες",
                         marker_color="skyblue"))

# Κορυφαία 5 χώρες με το μικρότερο προσδόκιμο ζωής για άνδρες
fig_men.add_trace(go.Bar(x=bottom_countries_men["country_name"],
                         y=bottom_countries_men["life_expectancy_men"],
                         name="Χώρες με το μικρότερο προσδόκιμο ζωής για άνδρες",
                         marker_color="lightblue"))

# Ρύθμιση τίτλου και άξονων
fig_men.update_layout(title="Προσδόκιμο Ζωής Ανδρών - Κορυφαία και Χαμηλότερα Κράτη",
                      xaxis_title="Χώρες",
                      yaxis_title="Προσδόκιμο Ζωής (Έτη)",
                      barmode="group",  # Ομαδοποίηση των μπαρ
                      height=600,
                      width=900)

fig_men.show()
```

```python
import plotly.express as px

# Υπολογισμός της διαφοράς προσδόκιμου ζωής (γυναίκες - άνδρες)
df_latest["life_expectancy_gap"] = df_latest["life_expectancy_women"] - df_latest["life_expectancy_men"]

# Απόκτηση των 10 χωρών με τη μεγαλύτερη και τη μικρότερη διαφορά προσδόκιμου ζωής
top_gap_countries = df_latest.nlargest(10, "life_expectancy_gap")
bottom_gap_countries = df_latest.nsmallest(10, "life_expectancy_gap")

# Συγχώνευση δεδομένων για το γράφημα
df_gap = pd.concat([top_gap_countries, bottom_gap_countries])

# Δημιουργία γραφήματος για τη διαφορά προσδόκιμου ζωής
fig_gap = px.bar(df_gap,
                 x="life_expectancy_gap",
                 y="country_name", 
                 color="life_expectancy_gap",
                 title=f"Διαφορά Προσδόκιμου Ζωής μεταξύ Γυναικών και Ανδρών ({latest_year})",
                 labels={"life_expectancy_gap": "Διαφορά (έτη)", "country_name": "Χώρα"},
                 height=600, 
                 orientation="h", 
                 color_continuous_scale="Viridis")

# Ρύθμιση του τίτλου και της διάταξης
fig_gap.update_layout(
    title_x=0.5,  # Κεντράρισμα του τίτλου
    xaxis_title="Διαφορά Προσδόκιμου Ζωής (Έτη)",
    yaxis_title="Χώρες",
    font=dict(family="Arial", size=12),  # Ρύθμιση γραμματοσειράς
    title_font=dict(size=18),  # Μέγεθος τίτλου
    coloraxis_showscale=True,  # Εμφάνιση κλίμακας χρώματος
    height=600,
    width=900,
    paper_bgcolor="rgba(245, 245, 245, 1)",  # Χρώμα φόντου
    plot_bgcolor="rgba(255, 255, 255, 1)",  # Χρώμα γραφήματος
)

# Εμφάνιση του γραφήματος
if kaggle:
    fig_gap.show(renderer='iframe_connected')
else:
    fig_gap.show()
```
